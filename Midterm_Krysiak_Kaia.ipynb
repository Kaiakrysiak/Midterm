```python
##### Question 2 #####

import pandas as pd

df = pd.read_csv('amazon-orders.csv')
df.head() 
```

```python
df.shape
```

```python
df = df.fillna(0)
df.head()
```

```python
df["Total Charged"] = df["Total Charged"].str.replace('$','').astype(float)
df.head()
```

```python
df["Total Charged"].sum()
```

```python
df["Total Charged"].mean()
```

```python
df["Total Charged"].median()
```

```python
df["Total Charged"].max()
```

```python
df["Total Charged"].min()
```

```python
df["Total Charged"].std()
```

```python
df["Tax Charged"] = df["Tax Charged"].str.replace('$','').astype(float)
df.head()
```

```python
df["Tax Charged"].sum()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df.head()
```

```python
df.plot.bar(x='Order Date', y='Total Charged', rot=90, figsize=(20,10))
```

```python
daily_orders = df.groupby('Order Date').sum()["Total Charged"]
daily_orders.head()
```

```python
df.plot.bar(x='Order Date', y='Tax Charged', rot=90, figsize=(20,10))
```

```python
daily_orders = df.groupby('Order Date').sum()["Tax Charged"]
daily_orders.head()
```
